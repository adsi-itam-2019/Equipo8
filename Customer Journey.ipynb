```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import collections as matcoll
```

```python
rtng=[4.2,6.4,7.3,5.4,6.5,9.5,8.5,2.2,7.7,8.5]
exp=['Bajar la aplicación',
     'Pantalla principal',
     'Filtrado de\n departamentos por\n características',
     'Busqueda en lista\n de departamentos',
     'Filtrado por precios',
     'Elección de\n departamentos',
     'Tratos con\n lugarteniente por\n mensajes o telefono',
     'Pago',
     'Llegada al departamento',
     'Contacto posterior\n con lugarteniente']
dic={'Event':exp,'Rating':rtng}
df=pd.DataFrame(data=dic).set_index('Event')
```

```python
plt.style.use('seaborn-white')

plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = 'Ubuntu'
plt.rcParams['font.monospace'] = 'Ubuntu Mono'
plt.rcParams['font.size'] = 10
plt.rcParams['axes.labelsize'] = 10
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['axes.titlesize'] = 10
plt.rcParams['xtick.labelsize'] = 8
plt.rcParams['ytick.labelsize'] = 8
plt.rcParams['legend.fontsize'] = 10
plt.rcParams['figure.titlesize'] = 12

width, height = plt.figaspect(0.4)
fig = plt.figure(figsize=(width,height), dpi=400)
ax = plt.subplot(111)
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
plt.ylim(0, 10)
#plt.xlim(0,9)

plt.yticks([])
#plt.xticks(fontsize=14)

plt.fill_between(x=(-0.2, 9.2),y1=0, y2=2.5, color='#FF8E8E')
plt.fill_between(x=(-0.2, 9.2),y1=2.5, y2=5, color='#FFD68E')
plt.fill_between(x=(-0.2, 9.2),y1=5, y2=7.5, color='#FFFF8E')
plt.fill_between(x=(-0.2, 9.2),y1=7.5, y2=10, color='#A8FF92')

plt.plot(df, linewidth=2, linestyle='-', marker='o', color='black')

#plt.axhline(y=2.5, color='#000000', linestyle=':', linewidth=0.5)
#plt.axhline(y=5, color='#000000', linestyle=':', linewidth=0.5)
#plt.axhline(y=7.5, color='#000000', linestyle=':', linewidth=0.5)
for i in range(0,len(exp)):
    plt.axvline(x=exp[i], ymin=0, ymax=(rtng[i]/10)-0.02, color='#000000', linestyle='--', linewidth=0.1)

# Axes alteration to put zero values inside the figure Axes
# Avoids axis white lines cutting through zero values - fivethirtyeight style
xmin, xmax, ymin, ymax = plt.axis()
plt.axis([xmin-0.1, xmax+0.1, ymin-0.1, ymax+0.4])

# Space plots a bit
plt.subplots_adjust(hspace=0.25, wspace=0.40)

plt.title("Customer Journey", fontsize=18)
plt.savefig('Customer Journey.png', bbox_inches='tight')
```
